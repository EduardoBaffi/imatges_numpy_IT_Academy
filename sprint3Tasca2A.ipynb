```python
##### - Exercici 1
# Crea un np.array d'una dimensió, que inclogui l'almenys 8 nombres sencers, data type int64. 
# Mostra la dimensió i la forma de la matriu.
import numpy as np

newArray = np.random.randint(0, 99,(12,),dtype='int64')
#newArray = np.array([ 80, 63,  8, 46, 88, 84, 58, 27, 53, 57,  6, 96],dtype='int64')

print(newArray)
print(newArray.dtype)
print("Forma de la matriz: " + str(newArray.shape))  # Comando alternativo print(np.shape(newArray))
print("Dimensión de la matriz: " + str(len(newArray.shape)))


# - Exercici 2
# De la matriu de l'exercici 1, calcula el valor mitjà dels valors introduïts i resta la mitjana resultant de cada un dels valors de la matriu.
print("Valor medio de los valores: " + str(np.mean(newArray)))
meanValue = np.mean(newArray)
subArray = newArray - meanValue #np.mean(newArray)
print(subArray) # Matriz con la media restada de cada un de los valores.

# - Exercici 3
# Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.
fiveArray = np.random.randint(0, 99,(5,5))
print(fiveArray)

maxValue = np.max(fiveArray)
print("Valor máximo de la matriz:"  + str(maxValue))

maxValueX = np.amax(fiveArray, axis=0)
print("El valor maximo del eje X es :")
print(maxValueX)

maxValueY = np.amax(fiveArray, axis=1)
print("El valor maximo del eje Y es :")
print(maxValueY)
```

```python
# - Exercici 4
#Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : 
# "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

import numpy as np

a = np.array([10, 15, 21, 26]) # Ejemplo 1 - Broadcasting posible porque tienen las mismas dimensiones
b = np.array([7, 9, 13, 17])

c = a * b
print(c)

d = np.random.randint(1, 10, (4,4)) # Ejemplo 2
e = 10
print(d)
print(d*e)

f = np.random.randint(1, 10, (4,4)) # Ejemplo 3
g = np.random.randint(1, 10, (4,1))
print("Matriz f:")
print(f)
print("f.shape = ", f.shape)
print("Matriz g:")
print(g)
print("g.shape = ", g.shape)
print("f + g")
print(f + g)

X = np.random.randint(1, 100, (5,5)) # Ejemplo 4
Y = np.random.randint(1, 10, (5,5))
print("X.shape = ", X.shape)
print("Y.shape = ", Y.shape)
Z = X*Y
print(Z)
print("Z=X*Y, Z.shape = : ", Z.shape)


# - Exercici 5
#Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.
indexArray = np.random.randint(1, 100, (6,6))
print(indexArray)
print("Columna 3 (index 2):")
print(indexArray[:,2])
print("Linea 5 (index 4):")
print(indexArray[4])

# - Exercici 6
# Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element...
# ... i comprovant si es divideix uniformement per quatre.

# Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

mask = indexArray %4 == 0
print(indexArray)
print(mask)

# - Exercici 7
#A continuació, utilitzeu aquesta màscara per indexar a la matriu de números original. 
#Això fa que la matriu perdi la seva forma original, reduint-la a una dimensió, però encara obteniu les dades que esteu cercant.

result = indexArray[mask]
print(result)
```

```python
# Manipulación de imágenes con matplotlib.

# - Ejercicio 8

# Mostreu-me a veure que passa quan eliminem el canal G Verd o B Blau.
# Hauries d'utilitzar la indexació per seleccionar el canal que voleu anul·lar.
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

image = mpimg.imread('cardona.jpg')  # Import original image
image.shape
print(image)
plt.show(plt.imshow(image))

imageNoRed = np.copy(image)  # Image without red 
imageNoRed[:, : , 0] = 0
plt.show(plt.imshow(imageNoRed))

imageNoGreen = np.copy(image) # Image without green
imageNoGreen[:, : , 1] = 0
plt.show(plt.imshow(imageNoGreen))

imageNoBlue = np.copy(image)  # Image without bue
imageNoBlue[:, : , 2] = 0
plt.show(plt.imshow(imageNoBlue))


# Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades...
#...i que haureu de pujar al vostre repositori a github.

# Saved images
mpimg.imsave('imageNoRed.jpg',imageNoRed)  
mpimg.imsave('imageNoGreen.jpg',imageNoGreen)
mpimg.imsave('imageNoBlue.jpg',imageNoBlue)
```
